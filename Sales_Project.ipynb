```python
#import libraries
import pandas as pd 
import seaborn as sns 
import glob 
import matplotlib.pyplot as plt 
```

```python
#Creat a list of all sales files as per month
sales_files=glob.glob('*.{}'.format('csv'))
sales_files
```

```python
#Combine data of all the sales files into one cvs file
# Method-I: Using concat function 
sales_df=pd.concat([pd.read_csv(f) for f in sales_files],ignore_index=True)
sales_df
```

```python
#Method -II: Using append function
sales_df.append([pd.read_csv(f) for f in sales_files],ignore_index=True)
sales_df 
```

```python
#Save combined data in new csv file
sales_df.to_csv('Combined_sales.csv')
```

```python
#Read combined file 
sales_df=pd.read_csv('Combined_sales.csv')
sales_df
```

```python
#Find out na values in the data
sales_df.isna()
```

```python
# Find out null values in the data
sales_df.isnull()
```

```python
#Clean the data by dropping na values
sales_df=sales_df.dropna()
```

```python
#Getting rid of text data in 'Order Date Column'
sales_df=sales_df[sales_df['Order Date'].str[0:2]!='Or']
```

```python
# Check information of data
sales_df.info()
```

```python
# Change datatype of 'Order Date' from object to datetime  
sales_df['Order Date']=pd.to_datetime(sales_df['Order Date'])
```

```python
sales_df.info()
```

```python
#Extract month from date and add month column in sales data
sales_df['Month']=sales_df['Order Date'].dt.month
sales_df['Month']
```

```python
sales_df
```

```python
#df_new.groupby(['Price Each','Purchase Address']).size().nlargest()
```

```python
# Extract city from purchase address and add city column into sales data
sales_df['City']=[(x.split(',')[1]) for x in (sales_df['Purchase Address'].values)]
sales_df['City']
```

```python
sales_df
```

```python
#Total Sales
sales_df['Price Each']=pd.to_numeric(sales_df['Price Each'])
sales_df['Quantity Ordered']=pd.to_numeric(sales_df['Quantity Ordered'])
sales_df['Total Sales']=sales_df['Quantity Ordered'] * sales_df['Price Each']
```

```python
# Month with maximum sale
sales_df.groupby('Month').sum('Total Sales').sort_values('Total Sales', ascending=False)
```

```python
# City with maximum product sold
City_max=sales_df.groupby(['City']).sum('Quantity Ordered').sort_values('Quantity Ordered',ascending=False)
City_max
```

```python
plt.plot(sales_df['Total Sales'], sales_df['Month'])
plt.show()
```
